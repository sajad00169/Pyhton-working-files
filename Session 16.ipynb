```python
d = {}

for i in unq_os:
    # print(i)
    v= ""
    bol = laptops["os"] == i
    os_v = laptops.loc[bol, "os_version"]
    bol1 = laptops.loc[bol, "os_version"].notnull()
    if os_v[bol1].shape[0]:
        v = os_v[bol1].iloc[0]
    else:
        v = None
    d[i] = v
```
