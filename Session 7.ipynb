```python
def convert_date(date):
    if date != "":
        date = date.replace("(","").replace(")","")
        date = int(date)
    return date
```

```python
bad_chr = ["(",")", "c", "C", ".", "'", "s", " "]


def demise(date):
    for c in bad_chr:
        date = date.replace(c,"")
    return date
```

```python
demise("(1960's)")
```

```python
for row in artworks:
    row[-2] = demise(row[-2])
```
